```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_file.csv"
student_data_to_load = "students_file.csv"

# Read School and Student Data File and store into Pandas Data Frames
schools_file = pd.read_csv(school_data_to_load)
students_file = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
all_ssdata = pd.merge(students_file, schools_file, how="left", on=["school_name", "school_name"])
all_ssdata.head()
```

```python
#Math pass rate
math_passrate = len(all_ssdata.loc[all_ssdata["math_score"] >= 70]) / len(all_ssdata["Student ID"])*100
math_passrate
```

```python
#read_passrate 
read_passrate = len(all_ssdata.loc[all_ssdata["reading_score"] >= 70]) / len(all_ssdata["Student ID"])*100
read_passrate
```

```python
#Calculate overall pass rate

overall_pass_rate = (math_passrate + read_passrate) /2
overall_pass_rate
```
